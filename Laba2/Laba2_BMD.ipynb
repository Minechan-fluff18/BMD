2. Побудова графіків. Matplotlib підтримує два методи побудови графіків: функціональний метод, який полягає у простій побудові графіків та їх елементарному перетворенні, об’єктно-орієнтований метод, який полягає у створенні графіків як об’єктів та подальшому виклику методів та атрибутів цього об’єкта.

2.1. Функціональний метод. Побудова базового графіку, що відображає залежність х від у.

```python
plt.plot(x, y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

Для побудови декількох графіків поруч:

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

2.2. Об’єктно-орієнтований метод. Етапи створення графіку:
- ініціалізуйте створення графіку як об’єкта;
- додайте до нього поле, для відображення графіка. В дужках задайте
параметри графіку. Параметри задаються у відсотковому значенні від 0 до 1 в
наступному порядку:
1 – відстань від 0 по осі х
2 – відстань від 0 по осі у
3 – ширина поля, для відображення графіку
4 – висота поля, для відображення графіку.

При умові створення лише одного графіку на полі, параметри
відповідають (0, 0, 1, 1)

```python
f=plt.figure()
axes=f.add_axes([0, 0, 1, 1])

axes.plot(x, y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

Для одночасної побудови декількох графіків:

```python
f=plt.figure()

axes1=f.add_axes([0, 0, 1, 1])
axes2=f.add_axes([0.1, 0.5, 0.4, 0.3])

axes1.plot(x, y, 'r')
axes1.set_xlabel('X label')
axes1.set_ylabel('Y label')
axes1.set_title('Title')

axes2.plot(y, x, 'b')
axes2.set_xlabel('X label')
axes2.set_ylabel('Y label')
axes2.set_title('Title')
```

Використання функції subplots при створенні графіка як об’єкта, його ініціалізація:

```python
f, axes=plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
f, axes=plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, 'r')
axes[1].plot(y, x, 'r')
```

```python
f, axes=plt.subplots(nrows=4, ncols=4)
```

```python
f, axes=plt.subplots(nrows=4, ncols=4)
plt.tight_layout()
```

Використання команди .legend()

```python
f=plt.figure()
a=f.add_axes([0, 0, 1, 1])

a.plot(x, x**2, label="x**2")
a.plot(x, x**3, label="x**3")
a.legend()
```

Зміна розміру графіку та DPI

```python
f=plt.figure(figsize=(8,8), dpi=50)
a=f.add_axes([0,0,1,1])

a.plot(x, x**2,label="x**2")
a.plot(x, x**3,label="x**3")
a.legend()
```

```python
f, a=plt.subplots()
a.plot(x,x**2,color='b')
a.plot(x,x**3,color='black')
```

```python
f, a=plt.subplots()
a.plot(x,x**2,lw=2, ls='--')
```

Задання типу, розміру та кольору маркерів:

```python
f, a=plt.subplots()
a.plot(x, x**2, marker='o', markersize=10, markerfacecolor='black')
```

Збереження побудованого графіку (savefig)
